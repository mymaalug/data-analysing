```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the dataset into a Pandas DataFrame
df = pd.read_csv('data.csv')
```

```python
# Dropping the duplicate rows
df = df.drop_duplicates()
```

```python
# Checking for null values
print(df.isnull().sum())
```

```python
fixed_acidity_mean = df['fixed acidity'].mean()
citric_acid_mean = df['citric acid'].mean()
chlorides_mean = df['chlorides'].mean()
free_sulfur_dioxide_mean = df['free sulfur dioxide'].mean()
total_sulfur_dioxide_mean = df['total sulfur dioxide'].mean()
density_mean = df['density'].mean()
pH_mean = df['pH'].mean()
quality_mean = df['quality'].mean()
```

##### Reason for replacing null values with the mean of each column
If you have a large dataset with many missing values, you can increase the accuracy of your analysis by filling them in with the mean of the appropriate column. This is due to the fact that the mean is a representative number that can provide a good estimate of the missing value.

```python
# Filling null values with mean
df['fixed acidity'] = df['fixed acidity'].fillna(fixed_acidity_mean)
df['citric acid'] = df['citric acid'].fillna(citric_acid_mean)
df['chlorides'] = df['chlorides'].fillna(chlorides_mean)
df['free sulfur dioxide'] = df['free sulfur dioxide'].fillna(free_sulfur_dioxide_mean)
df['total sulfur dioxide'] = df['total sulfur dioxide'].fillna(total_sulfur_dioxide_mean)
df['density'] = df['density'].fillna(density_mean)
df['pH'] = df['pH'].fillna(pH_mean)
df['quality'] = df['quality'].fillna(quality_mean)
```

```python
numerical_cols = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol', 'quality']
```

```python
# Create subplots for each column
fig, axs = plt.subplots(nrows=4, ncols=3, figsize=(15, 15))
# Flatten the axes to make it easier to iterate over them
axs = axs.flatten()
# Create histogram for each column
for i, col in enumerate(numerical_cols):
    plt.subplots_adjust(wspace=0.5, hspace=0.5)
    axs[i].hist(df[col], bins=10)
    axs[i].set_xlabel(col)
    axs[i].set_ylabel('Count')
# Remove any unused subplots
for i in range(len(numerical_cols), len(axs)):
    fig.delaxes(axs[i])
fig.suptitle('Histograms Of Red Wine Characteristics')
plt.show()
```

```python
#  A figure and an array of axis objects
fig, axs = plt.subplots(4, 3, figsize=(15, 15))
axs = axs.flatten()
for i, col in enumerate(numerical_cols):
    category_counts = df[col].value_counts()
    axs[i].pie(category_counts.values, labels=category_counts.index, autopct='%1.1f%%')
    axs[i].set_title(col)
    fig.suptitle('Distribution of Characteristics')
plt.show()
```

```python
fig, axs = plt.subplots(4, 3, figsize=(20, 20))
axs = axs.flatten()
for i, column in enumerate(numerical_cols):
    axs[i].bar(df[column].value_counts().index, df[column].value_counts().values)
    axs[i].set_title(column)
    fig.suptitle('Distribution of Characteristics')
plt.show()
```
